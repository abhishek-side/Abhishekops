```python
# training the k means cluster model
# optimim number of cluster = 5 because elbow graph shows less fall between 4 and 6
km=KMeans(n_clusters=5,init='k-means++',random_state=0)
# returns a label for each data point based on their cluster
Y=km.fit_predict(X)
Y
```

```python
# visualizing all the clusters
from matplotlib.pyplot import figure
plt.figure(figsize=(8,8),dpi=80)
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green',label='Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red',label='Cluster 1')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow',label='Cluster 1')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='blue',label='Cluster 1')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='violet',label='Cluster 1')

# plot the centroid
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=100, color='cyan', label='Centroids')
plt.title('Customers group')
plt.xlabel('Annual income$')
plt.ylabel('Spending score')
plt.show()
```
